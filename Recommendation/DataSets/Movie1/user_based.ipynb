```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
```

```python
ratings = pd.read_csv(r'C:\Users\adity\Documents\Machine Learning\Recommendation\DataSets\Movie1\ratings.csv')
ratings.head(10)
```

```python
movies = pd.read_csv(r'C:\Users\adity\Documents\Machine Learning\Recommendation\DataSets\Movie1\titles.csv')
movies.head(10)
```

```python
df = pd.merge(ratings, movies, how='left', on='movie_id')
df.head(10)
```

```python
movies_ratings = df.pivot_table(index='user_id', columns='movie_id', values='rating').fillna(0)
movies_ratings
```

```python
user_ratings = movies_ratings.T
user_ratings.head()
```

```python
scaler = StandardScaler()
ratings_scaled = scaler.fit_transform(user_ratings)
ratings_scaled
```

```python
movie_similarity = cosine_similarity(ratings_scaled)
movie_similarity
```

```python
similarity = pd.DataFrame(movie_similarity, index=user_ratings.index, columns=user_ratings.index)
similarity
```

```python
def recommend_movies(movie_id, num_recommendations=5):
    
    similarity_movie = similarity[movie_id].sort_values(ascending=False)
    
    similarity_movie = similarity_movie.drop(movie_id)
    
    top_recommendations = similarity_movie.head(num_recommendations)
    
    return top_recommendations
```
